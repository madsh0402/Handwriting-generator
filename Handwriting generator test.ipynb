# Handwriting Data processing

## Importing the strokes data

```python
import os
import xml.etree.ElementTree as ET

def parse_strokes(file_path):
    """
    Parses an XML file to extract stroke data.
    """
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    stroke_data = []
    for stroke in root.find('StrokeSet'):
        points = [(int(point.get('x')), int(point.get('y'))) for point in stroke]
        stroke_data.append(points)
    return stroke_data

def process_directory(directory_path):
    """
    Processes each directory to extract stroke data from XML files.
    """
    all_strokes = []
    for subdir, _, files in os.walk(directory_path):
        for file in files:
            if file.endswith('strokesz.xml'):
                file_path = os.path.join(subdir, file)
                strokes = parse_strokes(file_path)
                all_strokes.extend(strokes)
    return all_strokes

# Path to the 'original' directory
root_directory = 'Data/original'

# Process the directory and extract all strokes
strokes_data = process_directory(root_directory)

# Here you can add code to save 'strokes_data' to a file or process it further
```

### Plotting the strokes data

```python
import matplotlib.pyplot as plt

# Function to plot a single stroke
def plot_stroke(stroke):
    x, y = zip(*stroke)
    plt.plot(x, y, marker='.')

# Plotting the first few strokes to see what they look like
plt.figure(figsize=(10, 1))
for i, stroke in enumerate(strokes_data[:10]):
    plt.subplot(1, 10, i + 1)
    plot_stroke(stroke)
    plt.axis('equal')

plt.show()
```

These plots give us a glimpse into how the stroke data represents handwriting. While each plot shows only a single stroke, handwriting is typically composed of multiple such strokes combined. When these strokes are sequenced together, they form letters and words.

```python
# Plotting a sequence of strokes to see if it forms a more recognizable handwriting pattern
plt.figure(figsize=(12, 1))

# Combining the first few strokes into a single sequence
combined_strokes = [point for stroke in strokes_data[100:120] for point in stroke]
x_combined, y_combined = zip(*combined_strokes)

# Plotting the combined strokes
plt.plot(x_combined, y_combined, marker='.')
plt.title("Combined Strokes from First Few Sequences")
plt.axis('equal')
plt.show()
```
